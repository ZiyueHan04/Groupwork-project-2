## "Visualization of Average Annual Crime Trends (2020 to Present) in Los Angeles"

This code aims to visualize the average number of reported crimes per year in Los Angeles from 2020 to the present. The dataset is loaded, and the 'Date Rptd' column is converted to datetime format to facilitate temporal analysis. Subsequently, the data is filtered to include only incidents reported from 2020 onward. The script then groups the filtered data by year and calculates the average number of reported crimes for each year. Finally, a bar chart is generated, with each bar representing a year, and the height of each bar indicating the average number of reported crimes for that specific year. The visualization provides a clear overview of the annual crime trends, helping to identify any significant fluctuations or patterns in reported incidents over the specified time period.

```python
average_crimes_per_year = database.groupby(database['Date Rptd'].dt.year)['DR_NO'].count()

plt.figure(figsize=(10, 6))
plt.bar(average_crimes_per_year.index.astype(str), average_crimes_per_year.values, color='skyblue')
plt.title('Average Number of Reported Crimes per Year (2020 to Present)')
plt.xlabel('Year')
plt.ylabel('Average Number of Reported Crimes')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Analysis 1: Visualize the trend of reported crimes over the years

```python
database['Year'] = database['Date Rptd'].dt.year
database['Month'] = database['Date Rptd'].dt.month
crime_trend_over_years = database.groupby('Year')['DR_NO'].count()
plt.figure(figsize=(12, 6))
plt.plot(crime_trend_over_years, marker='o')
plt.title('Trend of Reported Crimes Over the Years (2020 to Present)')
plt.xlabel('Year')
plt.ylabel('Number of Reported Crimes')
plt.grid(True)
plt.show()
```

Analysis 2: Distribution of crimes by month

```python
database['Year'] = database['Date Rptd'].dt.year
database['Month'] = database['Date Rptd'].dt.month
crime_distribution_by_month = database.groupby('Month')['DR_NO'].count()
plt.figure(figsize=(10, 6))
plt.bar(crime_distribution_by_month.index, crime_distribution_by_month.values, color='skyblue')
plt.title('Distribution of Reported Crimes by Month')
plt.xlabel('Month')
plt.ylabel('Number of Reported Crimes')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
database['Year'] = database['Date Rptd'].dt.year
database['Month'] = database['Date Rptd'].dt.month
top_crime_types = database['Crm Cd Desc'].value_counts().nlargest(10)
plt.figure(figsize=(12, 6))
top_crime_types.plot(kind='bar', color='coral')
plt.title('Top 10 Crime Types')
plt.xlabel('Crime Type')
plt.ylabel('Number of Reported Crimes')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Question 1: Temporal Variability

Explanation: This code visualizes the monthly distribution of reported crimes over the years, demonstrating how it varies.

```python
database['Year'] = database['Date Rptd'].dt.year
database['Month'] = database['Date Rptd'].dt.month

monthly_crime_distribution = database.groupby(['Year', 'Month'])['DR_NO'].count()
fig, ax = plt.subplots(figsize=(12, 6))
for year in monthly_crime_distribution.index.levels[0]:
    monthly_data = monthly_crime_distribution.loc[year]
    ax.plot(monthly_data.index, monthly_data.values, label=str(year))

ax.set_title('Monthly Distribution of Reported Crimes Over the Years')
ax.set_xlabel('Month')
ax.set_ylabel('Number of Reported Crimes')
ax.legend()
plt.show()
```

Question 2: Day-of-Week Patterns

Explanation: This code analyzes the average number of reported crimes for each day of the week over different years, then visualizes the pattern using a fan chart.

```python
database['Year'] = database['Date Rptd'].dt.year
database['DayOfWeek'] = database['Date Rptd'].dt.dayofweek

daily_crime_distribution = database.groupby(['Year', 'DayOfWeek'])['DR_NO'].count()

fig, ax = plt.subplots(figsize=(12, 6))
for year in daily_crime_distribution.index.levels[0]:
    daily_data = daily_crime_distribution.loc[year]
    ax.plot(daily_data.index, daily_data.values, label=str(year))

ax.set_title('Day-of-Week Patterns in Reported Crimes Over the Years')
ax.set_xlabel('Day of the Week')
ax.set_ylabel('Number of Reported Crimes')
ax.legend()
plt.show()
```

Seasonal Crime Patterns


```python
database['Month'] = database['Date Rptd'].dt.month
monthly_crime_counts = database.groupby('Month')['DR_NO'].count()

plt.figure(figsize=(10, 6))
plt.plot(monthly_crime_counts.index, monthly_crime_counts.values, marker='o')
plt.title('Seasonal Crime Patterns')
plt.xlabel('Month')
plt.ylabel('Number of Reported Crimes')
plt.show()
```

Top 10 High-Crime Areas Over Time

```python
average_crime_by_area = database.groupby(['AREA NAME', 'Year'])['DR_NO'].count().reset_index()
top_10_areas = average_crime_by_area.groupby('AREA NAME')['DR_NO'].mean().nlargest(10).index

top_10_areas_data = average_crime_by_area[average_crime_by_area['AREA NAME'].isin(top_10_areas)]

plt.figure(figsize=(12, 8))
for area in top_10_areas_data['AREA NAME'].unique():
    area_data = top_10_areas_data[top_10_areas_data['AREA NAME'] == area]
    plt.plot(area_data['Year'], area_data['DR_NO'], label=area)

plt.title('Top 10 High-Crime Areas Over Time')
plt.xlabel('Year')
plt.ylabel('Average Number of Reported Crimes')
plt.legend()
plt.show()
```

Weapon Usage Trends

```python
weapon_usage_data = database[['Year', 'Weapon Used Cd', 'Weapon Desc']]
weapon_counts = weapon_usage_data.groupby(['Year', 'Weapon Desc'])['Weapon Used Cd'].count().reset_index()
weapon_counts_pivot = weapon_counts.pivot(index='Year', columns='Weapon Desc', values='Weapon Used Cd').fillna(0)
weapon_counts_pivot.plot(kind='area', stacked=True, figsize=(12, 8))
plt.title('Weapon Usage Trends Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Reported Crimes')
plt.show()
```

Grouping the data by year and weapon description. To enhance visual clarity, the data is pivoted to create a stacked area chart where each color represents a distinct weapon category.

Users can discern fluctuations in the prevalence of specific weapons, potentially identifying trends or shifts in criminal behavior. This visual representation facilitates a nuanced understanding of the dynamics of crime, aiding law enforcement, policymakers, and analysts in developing targeted strategies to address weapon-related incidents.
